```python
# BERT는 이미 누군가가 학습해둔 모델을 사용하는 것이므로 우리가 사용하는 모델과 토크나이저는 항상 맵핑 관계여야 합니다.
from transformers import TFBertForMaskedLM
from transformers import AutoTokenizer
```

```python
model = TFBertForMaskedLM.from_pretrained('bert-large-uncased') # 언어 모델 로드
tokenizer = AutoTokenizer.from_pretrained("bert-large-uncased") # 토크나이저 로드
```

```python
inputs = tokenizer('Soccer is a really fun [MASK].', return_tensors='tf')
print(inputs['input_ids'])
```

```python
print(inputs['attention_mask']) # 어텐션 마스크
```

### FillMaskPipeline

모델과 토크나이저 넣으면 [MASK] 토큰 예측하는 내용들 다 나옴

```python
from transformers import FillMaskPipeline
pip = FillMaskPipeline(model=model, tokenizer=tokenizer)
```

```python
pip('Soccer is a really fun [MASK].')
```

```python
pip('The Avengers is a really fun [MASK].')
```
